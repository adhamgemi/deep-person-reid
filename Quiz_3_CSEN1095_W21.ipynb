# Question 1 : Data Tidying



##Observe the following dataset: https://raw.githubusercontent.com/raneemsultan/Quiz3/main/disease.csv  and use data tidying techniques to make the dataset tidy. (5 Marks)

### Make sure you apply the necessary text transformations on your columns. Your final dataframe should contain meaningful colum values and no duplicate columns.


```python
import pandas as pd 
disease_df = pd.read_csv("https://raw.githubusercontent.com/raneemsultan/Quiz3/main/disease.csv")
disease_df
```

```python
disease_df_cases = disease_df.loc[:,'Cases_Guinea':'Cases_Mali']
```

```python
disease_df_cases
```

```python
disease_df_cases =pd.melt(disease_df_cases)
disease_df_cases.head()
```

```python
disease_df_cases =disease_df_cases.rename(columns={'value':'cases Value'})
disease_df_cases.head()
```

```python
#
disease_df_cases['Countries'] = disease_df_cases['variable'].str.extract('(\d+)')
disease_df_cases['Countries']
```

```python
disease_df_cases = disease_df_cases.drop('variable',axis=1)
```

```python
disease_df_deaths = pd.melt(disease_df.loc[:,'Deaths_Guinea':'Deaths_Mali'])
disease_df_deaths.head()
```

```python
disease_df_deaths['Countries'] = disease_df_deaths['variable'].str.extract('(\d+)')
```

```python
disease_df_deaths.head()
```

```python
disease_df_deaths = disease_df_deaths.drop('variable',axis=1)
```

```python
disease_df_deaths
```

```python
disease_df_deaths = disease_df_deaths.rename(columns={'value':'deaths count'})
disease_df_deaths
```

```python
disease_df_final = pd.melt(disease_df.loc[:,'Date':'Cases_Guinea'],id_vars=['Date','Day'])#get the dataframe from the starting column up until gdpPercap_2007 and melt it 
disease_df_final
```

```python
disease_df_final = disease_df_final.drop(['variable','value'],axis=1) #drop the columns var and val as they will be replaced by what has been created from the pop and gdp dfs
disease_df_final 
```

```python
disease_df_final = pd.concat([disease_df_final,disease_df_deaths,disease_df_cases.loc[:,['cases count']]],axis=1)#merge the dataframes  tiegther 
disease_df_final
```
